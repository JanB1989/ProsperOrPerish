# Goods × Attributes Output Modifier Matrix

This notebook loads and displays the output modifier matrix for goods production by location attributes. The matrix is stored in `data/goods_output_modifiers.xlsx` and can be edited in Excel. Per-cell bounds: [-1.0, 0.25]. The final modifier for a location is the sum across applicable attributes (one topography + one vegetation + one climate + water bonuses if present).

```python
import pandas as pd

from IPython.display import display
from analysis.building_levels.building_analysis import load_goods_output_modifiers, get_path

df = load_goods_output_modifiers()
print(f"Matrix shape: {df.shape[0]} goods x {df.shape[1]} attributes")
df
```

## Matrix by Attribute Group

Columns grouped by Topography, Vegetation, Climate, and Water.

```python
# Column groups (from generate_goods_modifier_matrix.py) - buildable topographies only
topography = ["flatland", "hills", "plateau", "mountains", "wetlands", "salt_pans", "atoll"]
vegetation = ["desert", "sparse", "grasslands", "farmland", "woods", "forest", "jungle"]
climate = ["tropical", "subtropical", "oceanic", "arid", "cold_arid", "mediterranean", "continental", "arctic"]
water = ["has_river", "is_adjacent_to_lake", "is_coastal"]

for name, cols in [("Topography", topography), ("Vegetation", vegetation), ("Climate", climate), ("Water", water)]:
    valid = [c for c in cols if c in df.columns]
    if valid:
        display(df[valid].style.set_caption(name))
```
